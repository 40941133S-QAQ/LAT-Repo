```python
#Q1：求各欄位名稱
# 讀取第一列標題，並輸出所有欄位名稱
column_names = df.columns.tolist()
print(column_names)
```

```python
#Q2：各欄位有什麼內容
# 輸出每個欄位的 features
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Column '{column}' has {len(unique_values)} unique values:\n{unique_values}\n")
```

```python
#區分人數、學校數的表格
data_人=df.copy()
data_校=df.copy()
data_人= data_人[data_人['StatCourseName'] != '高級中等學校校數及教職員數─按行政區別分-校數(單位:所)']
data_校= data_校[data_校['StatCourseName'] == '高級中等學校校數及教職員數─按行政區別分-校數(單位:所)']
```

```python
#Q3：此資料統計多少學校
num_校 = data_校['FValue'].sum()
print("此資料統計",num_校,"間學校")
```

```python
#Q4：此資料統計多少人
num_人 = data_人['FValue'].sum()
print("此資料統計",num_人,"人")
```

```python
#Q5：男女各多少人
data_人.groupby(['Complex2Name'])['FValue'].sum()
```

```python
#Q6：教師、職員各多少人
data_人.groupby(['StatCourseName'])['FValue'].sum()
```

```python
#Q7：各區教職員男女人數
sex_by_place = data_人.groupby(['PlaceName','Complex2Name'])['FValue'].sum().unstack()
sex_by_place
```

```python
#Q8：計算各區教職員男女比例
sex_by_place['女比例']=sex_by_place['女']/sex_by_place.sum(axis=1)
sex_by_place['男比例']=sex_by_place['男']/sex_by_place.sum(axis=1)
sex_by_place
```

```python
#Q9計算各區學校國/市/私立校數
temp = data_校.groupby(['PlaceName','ComplexName'])['FValue'].sum().unstack()
temp
```

```python
#Q10：哪區老師最多
num= data_人.groupby(['PlaceName'])['FValue'].sum().reset_index()
num.columns = ['PlaceName', 'number']
num
max_num= num['number'].max() #最大值
max_index= num['number'].idxmax() #存儲其所在行的索引
max_num_PlaceName=num.loc[max_index]['PlaceName'] #使用所找到的索引在'PlaceName'列中找到相應的值
print(max_num_PlaceName,'老師人數最多' )
```
